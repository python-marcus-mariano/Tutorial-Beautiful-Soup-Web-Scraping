## White House Example

YouTube Link:

**Let's obtain the links from the following website: [here.](https://www.whitehouse.gov/briefings-statements/)** 

One of the things this website consists of is records of presidential
briefings and statements.

Goal: Extract all of the links on the page that point to the 
briefings and statements.


```python
from requests import get
from bs4 import BeautifulSoup as bs4
```

```python
result = get("https://www.whitehouse.gov/briefings-statements/")
src = result.content
soup = bs4(src, 'lxml')
```

```python
soup
```

```python
urls = []
for h2_tag in soup.find_all('h2'):
    a_tag = h2_tag.find('a')
    urls.append(a_tag.attrs['href'])
```

```python
urls
```

```python
time_data = []
for p_tag in soup.find_all('p'):
    if p_tag.find('time'):        
        time_tag = p_tag.find('time')    
        time_data.append(time_tag)
```
